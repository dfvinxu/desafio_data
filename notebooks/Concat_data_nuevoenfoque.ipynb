```python
import pandas as pd
```

```python
df_fuentes = pd.read_csv("../data/madrid/fuentes/fuentes_nuevo.csv")
df_museos = pd.read_csv("../data/madrid/museos/museos_nuevo.csv")
df_parques = pd.read_csv("../data/madrid/Parques y jardines/parques_y_jardines_nuevo.csv")
df_oficinas = pd.read_csv("../data/madrid/oficinas_turismo/tourist_office_nuevo.csv")
df_piscinas = pd.read_csv("../data/madrid/Piscinas/piscinas_madrid_nuevo.csv")
```
